```
import matplotlib.pyplot as plt
import numpy as np
def cyclicLR(num_iterations,base,max,step_size) :
        lines = []
        for iteration in range(num_iterations) :
            cycle = np.floor(1+iteration/(2*step_size))
            x = np.abs(iteration/step_size-2*cycle +1)
            line = base + (max-base)*(np.maximum(0,1-x))
            lines.append(line)
        plt.plot(list(range(num_iterations)),lines)
cyclicLR(num_iterations=10000,base=0.001,max=0.006,step_size=1000)
```
